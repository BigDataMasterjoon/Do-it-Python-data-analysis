```python
# 분석 도전
midwest = pd.read_csv('midwest.csv')
```

```python
midwest_new = midwest.assign(popminors_p = lambda x: (x['poptotal'] - x['popadults']) / x['poptotal'] * 100)
midwest_new
```

```python
midwest_new[['county', 'popminors_p']].sort_values('popminors_p', ascending= False) \
                                      .head()
```

```python
midwest_new = midwest_new.assign(popminors_grade = np.where(midwest_new['popminors_p'] >= 40, 'large',
                                                   np.where(midwest_new['popminors_p'] >= 30, 'middle', 'small')))
midwest_new
```

```python
midwest_new.groupby('popminors_grade')\
           .agg(n = ('popminors_grade', 'count'))
```

```python
# 4
midwest_new.assign(popasian_p = lambda x: x['popasian'] / x['poptotal'] * 100) \
           [['state', 'county', 'popasian_p']] \
           .sort_values('popasian_p') \
           .head(10)
```
