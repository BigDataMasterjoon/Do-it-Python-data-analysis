```python
# total 기준으로 A, B, C 등급부여
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
                        np.where(mpg['total'] >= 20, 'B', 'C'))

# 데이터 확인
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
# 등급 빈도표 알파벳순 정렬
count_grade = mpg['grade'].value_counts().sort_values()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
# 2
midwest = midwest_raw.copy()
midwest = midwest.rename(columns= {'poptotal' : 'total',
                                   'popasian' : 'asian'})
midwest.info()
```

```python
# 3
midwest['asian/all'] = midwest['asian'] / midwest['total']
midwest
```

```python
# 히스토그램
midwest['asian/all'].plot.hist()
```

```python
# 4
midwest['asian/all'].mean()
```

```python
criteria = midwest['asian/all'].mean()
```

```python
midwest['asian_grade'] = np.where(midwest['asian/all']> criteria, 'large', 'small')
midwest
```

```python
midwest['asian_grade'].value_counts()
```

```python
midwest['asian_grade'].value_counts().plot.bar(rot = 0)
```
