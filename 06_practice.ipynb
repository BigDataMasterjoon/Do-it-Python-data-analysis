```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
```

```python
# x축은 displ, y축은 hwy를 나타낸 산점도 만들기
import seaborn as sns
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy')
```

```python
# x축 범위 3~6으로 제한
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy') \
   .set(xlim = [3, 6])
```

```python
# x축 범위 3~6, y축 범위 10~30 제한
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy') \
   .set(xlim = [3, 6], ylim = [10, 30])
```

```python
# drv별로 표식 색깔 다르게 표현
sns.scatterplot(data = mpg, x = 'displ', y = 'hwy', hue = 'drv')
```

```python
df_mpg = mpg.groupby('drv', as_index= False) \
            .agg(mean_hwy = ('hwy', 'mean'))
df_mpg
```

```python
sns.barplot(data = df_mpg, x = 'drv', y = 'mean_hwy')
```

```python
# 데이터 프레임 정렬하기
df_mpg = df_mpg.sort_values('mean_hwy', ascending= False)

# 막대 그래프 만들기
sns.barplot(data = df_mpg, x = 'drv', y = 'mean_hwy')
```

```python
df_mpg = mpg.groupby('drv', as_index= False) \
            .agg(n = ('drv', 'count'))
df_mpg
```

```python
# 막대 그래프 만들기
sns.barplot(data = df_mpg, x = 'drv', y = 'n')
```

```python
# 빈도 막대 그래프 만들기
sns.countplot(data = mpg, x = 'drv')
```

```python
# economics 데이터 불러오기
import pandas as pd
economics = pd.read_csv('economics.csv')
economics.head()
```

```python
import seaborn as sns
sns.lineplot(data = economics, x = 'date', y = 'unemploy')
```

```python
# 날짜 시간 타입 변수 만들기
economics['date2'] = pd.to_datetime(economics['date'])
economics
```

```python
# 변수 타입 확인
economics.info()
```

```python
# 연추출
economics['date2'].dt.year
```

```python
# 월 추출
economics['date2'].dt.month
```

```python
# 일 추출
economics['date2'].dt.day
```

```python
# 연도 변수 추가
economics['year'] = economics['date2'].dt.year
economics.head()
```

```python
# x축에 연도 표시
sns.lineplot(data = economics, x = 'year', y = 'unemploy')
```

```python
# 혼자서 해보기
economics.head()
```

```python
sns.lineplot(data = economics, x = economics['year'], y = 'psavert', errorbar= None)
```

```python
economics['month'] = economics['date2'].dt.month
economics.head()
```

```python
df = economics.query('year == 2014')
sns.lineplot(data = df, x = 'month', y = 'psavert')
```
