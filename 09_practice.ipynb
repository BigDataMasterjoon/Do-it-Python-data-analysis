```python
# pip install kiwipiepy
from kiwipiepy import Kiwi

text = "대한민국의 영토는 한반도와 그 부속도서로 한다"

kiwi = Kiwi()
tokens = kiwi.tokenize(text)

[token.form for token in tokens if token.tag.startswith("N")]
```

```python
# 연설문에서 명사 추출하기

kiwi = Kiwi()
tokens = kiwi.tokenize(moon)

nouns = [token.form for token in tokens if token.tag.startswith("N")]
nouns
```

```python
# 데이터 프레임으로 변환
import pandas as pd
df_word = pd.DataFrame({'word': nouns})
df_word
```

```python
# 글자 수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
# 단어 빈도 구하기
df_word = df_word.groupby('word', as_index= False) \
                 .agg(n = ('word', 'count')) \
                 .sort_values('n', ascending= False)
df_word
```

```python
# 단어 빈도 상위 20개 추출
top_20 = \
    df_word.head(20)
top_20
```

```python
# 데이터 불러오기
import pandas as pd
df = pd.read_csv('news_comment_BTS.csv', encoding= 'UTF-8')

# 데이터 살펴보기
df.info()
```

```python
df.head()
```

```python
# 불필요한 문자 제거하기
df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True)
df['reply'].head()
```

```python
# pip install kiwipiepy
from kiwipiepy import Kiwi

kiwi = Kiwi()

def extract_nouns(text):
    if not isinstance(text, str):
        return []
    tokens = kiwi.tokenize(text)  # Token 객체 리스트 반환
    return [t.form for t in tokens if t.tag.startswith('NN')]  # NNG/NNP 등 명사만

# pandas Series에 적용
nouns = df['reply'].apply(extract_nouns)

nouns
```

```python
pd.Series(['안녕', '좋아요']).apply(len)
```

```python
nouns = nouns.explode()
nouns
```

```python
# 데이터 프레임 만들기
df_word = pd.DataFrame({'word' : nouns})

# 글자 수 추가
df_word['count'] = df_word['word'].str.len()

# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word
```

```python
# 빈도표 만들기
df_word = df_word.groupby('word', as_index= False) \
                 .agg(n = ('word', 'count')) \
                 .sort_values('n', ascending= False)
df_word
```

```python
# 단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

```python
# 가로 세로 크기 설정
plt.rcParams.update({'figure.figsize' : [6.5, 6]})

# 막대 그래프 만들기
sns.barplot(data = top20, y = 'word', x = 'n')
```

```python
# 워드클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드클라우드 출력하기
plt.figure(figsize= (10, 10)) # 가로 세로 크기 설정
plt.axis('off')               # 테두리 선 없애기
plt.imshow(img_wordcloud)     # 워드 클라우드 출력
```
